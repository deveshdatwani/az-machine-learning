```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('data/50_Startups.csv')
```

```python
dataset
```

```python
new_york = dataset[dataset['State'] == 'New York']
```

```python
r_d = new_york.iloc[:,0].values
marketting_spend = new_york.iloc[:,2:3].values
```

```python
newyork_profit = new_york.iloc[:,-1:].values
newyork_profit
```

```python
from matplotlib import pyplot as plt
```

## Visualizing co-relation between r&d spend, marketing spend and profits

```python
plt.scatter(r_d, newyork_profit, color="red")
plt.scatter(marketting_spend, newyork_profit, color='blue')
plt.xlabel('R&D/Marketing spend')
plt.ylabel('Profit')
plt.title('RnD vs Profit in New York')
plt.show() 
```

```python
plt.scatter(marketting_spend, newyork_profit, color="red")
plt.xlabel('Marketting spend')
plt.ylabel('Profit')
plt.title('Marketting vs Profit in New York')
plt.show() 
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1:].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
c_transformer = ColumnTransformer(transformers=[('lol',OneHotEncoder(),[3])],remainder='passthrough')
X = np.array(c_transformer.fit_transform(X))
X1 = c_transformer.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_prediction = regressor.predict(X_test)
```

```python
from matplotlib import pyplot as plt
plt.scatter(range(len(y_prediction)), y_prediction, color='red')
plt.scatter(range(len(y_prediction)), y_test, color='blue')
plt.title('multiple varibale regression')
plt.show() 
```

```python
y_prediction_train = regressor.predict(X_train)
```

```python
plt.scatter(range(len(y_prediction_train)), y_prediction_train, color='red')
plt.scatter(range(len(y_prediction_train)), y_train, color='blue')
plt.title('multiple varibale regression')
plt.show() 
```
