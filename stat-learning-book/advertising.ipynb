```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data_set = pd.read_csv('data/Advertising.csv')
print(data_set.describe())
```

```python
tv_expenditure = data_set['TV']
sales = data_set['sales']
```

```python
x_val = [i for i in range(len(data_set))]
plt.scatter(tv_expenditure, sales)
plt.show()
```

```python
radio_expenditure = data_set['radio']
plt.scatter(radio_expenditure, sales)
```

```python
radio_expenditure = data_set['radio']
plt.scatter(radio_expenditure, sales)
plt.show()
```

```python
x_data = tv_expenditure.values
y = sales
```

```python
from sklearn.linear_model import LinearRegression as lg_model
```

```python
x_data = np.reshape(x_data, (200,1))
simple_model = lg_model()
simple_model.fit(x_data, y)
```

```python
print(simple_model.coef_)
print(simple_model.intercept_)
```

```python
y_pred = simple_model.predict(x_data)
```

```python
plt.scatter(x_data, y_pred, color='red')
plt.scatter(x_data, y, color='blue')
plt.show() 
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error as mse
print('r2 score is {}'.format(r2_score(y, y_pred)))
print('mse score is {}'.format(mse(y, y_pred)))
```
